# Yelp Reviews MVP

Connie Xiao

Restaurants from Boston were extracted from the Yelp dataset. There are 2846 restaurants with over 500,000 reviews altogether. All reviews that rated 4 or more stars were assigned a positive sentimental value while 3 star was assigned to a neutral value and everything else as a negative value. Then I sampled 20,000 reviews from each sentiment to a new dataframe, giving me 60,000 reviews to work with. Reviews were cleaned and stop words were removed. First a TF-IDF matrix was made with terms and topic, then I wanted to see the top words for each topic.

This is a good starting point for the baseline. There are topics such as japanese cuisine, american cuisine, pizza, and seafood. Dimensionality reduction with SVD will be needed to get rid of irrelevant terms. After unsupervised learning is done, a supervised classification will be done to see if reviews can predict the rating star.
